```python
import warnings
warnings.filterwarnings(
    "ignore",
    message="Was asked to gather along dimension 0, but all input tensors were scalars"
)

import shutil
from pathlib import Path

# Clear Hugging Face cache to ensure fresh downloads
print("üßπ Cleaning Hugging Face cache...")
cache_dir = Path("/root/.cache/huggingface")
if cache_dir.exists():
    shutil.rmtree(cache_dir)
print("Cache cleared.\n")

# Install required libraries
print("üì¶ Installing libraries...")
!pip install -U transformers datasets evaluate accelerate matplotlib seaborn scikit-learn -q
print("Libraries installed.\n")

import os
import time
import numpy as np
import pandas as pd
import torch
import evaluate
import matplotlib.pyplot as plt
import seaborn as sns

from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
    DataCollatorWithPadding
)
from datasets import Dataset, ClassLabel
from sklearn.metrics import f1_score, confusion_matrix, precision_score, recall_score

sns.set_theme(style="whitegrid")
plt.rcParams.update({'figure.figsize': (12, 7), 'font.size': 12})

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# List of models to evaluate
model_list = [
    "answerdotai/ModernBERT-base",
    "distilroberta-base",
    "distilbert-base-uncased",
    "bert-base-uncased",
    "roberta-base"
]

# Data structures to store evaluation results
summary_metrics = []
per_class_metrics_list = []
per_class_precision_list = []
per_class_recall_list = []
confusion_matrices = {}
learning_curve_data = {}

# Helper function to measure model size
def measure_model_size(model):
    """Calculates model size in MB."""
    temp_dir = Path("temp_model_size_check")
    temp_dir.mkdir(exist_ok=True)
    try:
        model.save_pretrained(temp_dir)
        total_size = sum(f.stat().st_size for f in temp_dir.iterdir() if f.is_file())
        return total_size / (1024 * 1024) # MB
    finally:
        if temp_dir.exists():
            shutil.rmtree(temp_dir)

# Helper function to measure inference time
def measure_inference_time(model, tokenizer, texts, device, num_warmup=2, num_runs=10):
    """Measures average inference time per sentence in milliseconds."""
    model.eval()
    inputs = tokenizer(texts, return_tensors="pt", padding=True, truncation=True, max_length=128).to(device)

    # Warmup runs to stabilize measurements
    for _ in range(num_warmup):
        with torch.no_grad(): _ = model(**inputs)

    # Measure inference time
    start_time = time.time()
    for _ in range(num_runs):
        with torch.no_grad(): _ = model(**inputs)
    end_time = time.time()

    avg_time_seconds = (end_time - start_time) / num_runs
    return (avg_time_seconds / len(texts)) * 1000 # ms per sentence

# Attempt to load the dataset from various paths
potential_paths = [
    "/content/final_dataset.csv",
    "/content/data/final_dataset.csv",
    "/kaggle/input/mentalhealth/final_dataset.csv",
    "finalized.csv"
]

dataset_path = next((p for p in potential_paths if os.path.exists(p)), None)

# Load dataset or create dummy data if not found
if not dataset_path:
    print("‚ö†Ô∏è Dataset not found. Using dummy data.")
    df = pd.DataFrame({
        "text": ["I feel anxious and stressed"] * 50 +
                ["Everything feels fine"] * 50 +
                ["I am completely burned out"] * 50 +
                ["Depression is exhausting"] * 50,
        "label": ["anxiety", "normal", "burnout", "depression"] * 50
    })
else:
    print(f"‚úÖ Dataset loaded from: {dataset_path}")
    df = pd.read_csv(dataset_path)

# Preprocess the dataset
df.dropna(subset=["text"], inplace=True)
df = df[df["text"].str.split().str.len() > 2].copy()

# Sample texts for inference measurement
sample_texts_for_inference = df['text'].sample(min(50, len(df)), random_state=42).tolist()

# Plot class distribution
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x="label")
plt.title("Class Distribution in Final Dataset")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()

# Map labels to integers for model training
label_list = sorted(df["label"].unique())
label_map = {l: i for i, l in enumerate(label_list)}
reverse_label_map = {i: l for l, i in label_map.items()}
df["labels"] = df["label"].map(label_map)

# Convert DataFrame to Hugging Face Dataset format
dataset = Dataset.from_pandas(df[["text", "labels"]])
dataset = dataset.cast_column("labels", ClassLabel(names=label_list))

# Determine optimal max_length for tokenization
print("\nüîç Computing optimal max_length...")
temp_tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
token_lengths = [len(temp_tokenizer.encode(t)) for t in df["text"]]
max_length = int(min(np.percentile(token_lengths, 95), 512))
print(f"Using max_length = {max_length}\n")

# Function to freeze lower layers of the model
def freeze_lower_layers(model, freeze_ratio=0.7):
    encoder_layers = None
    if hasattr(model, "bert"): encoder_layers = model.bert.encoder.layer
    elif hasattr(model, "roberta"): encoder_layers = model.roberta.encoder.layer
    elif hasattr(model, "distilbert"): encoder_layers = model.distilbert.transformer.layer
    elif hasattr(model, "base_model") and hasattr(model.base_model, "layer"): encoder_layers = model.base_model.layer

    if encoder_layers:
        freeze_until = int(len(encoder_layers) * freeze_ratio)
        for layer in encoder_layers[:freeze_until]:
            for param in layer.parameters():
                param.requires_grad = False

# Main loop for model training and evaluation
for model_name in model_list:
    print(f"\n{'='*60}\nüöÄ MODEL: {model_name}\n{'='*60}")

    # Initialize tokenizer and set pad token if necessary
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token

    # Tokenize the dataset
    def tokenize(batch):
        return tokenizer(batch["text"], truncation=True, max_length=max_length)

    tokenized = dataset.map(tokenize, batched=True)
    # Split dataset into training and testing sets
    split = tokenized.train_test_split(test_size=0.2, seed=42, stratify_by_column="labels")

    # Load model for sequence classification
    model_kwargs = {"num_labels": len(label_list)}
    if "ModernBERT" in model_name:
        model_kwargs["reference_compile"] = False

    model = AutoModelForSequenceClassification.from_pretrained(model_name, **model_kwargs).to(device)
    model.config.pad_token_id = tokenizer.pad_token_id
    freeze_lower_layers(model)

    # Load evaluation metrics
    metric_acc = evaluate.load("accuracy")
    metric_f1 = evaluate.load("f1")

    # Function to compute metrics for evaluation
    def compute_metrics(eval_pred):
        logits, labels = eval_pred
        preds = np.argmax(logits, axis=-1)
        return {
            "accuracy": metric_acc.compute(predictions=preds, references=labels)["accuracy"],
            "f1_weighted": metric_f1.compute(predictions=preds, references=labels, average="weighted")["f1"],
            "f1_macro": metric_f1.compute(predictions=preds, references=labels, average="macro")["f1"],
            "precision_macro": precision_score(labels, preds, average="macro", zero_division=0),
            "recall_macro": recall_score(labels, preds, average="macro", zero_division=0)
        }

    # Configure training arguments
    training_args = TrainingArguments(
        output_dir=f"./results_{model_name.replace('/', '_')}",
        num_train_epochs=3,
        per_device_train_batch_size=32 if "distil" in model_name.lower() else 16,
        per_device_eval_batch_size=64,
        warmup_ratio=0.1,
        weight_decay=0.01,
        eval_strategy="epoch",
        save_strategy="no",
        fp16=torch.cuda.is_available(),
        logging_steps=50,
        report_to="none"
    )

    # Initialize and run Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=split["train"],
        eval_dataset=split["test"],
        compute_metrics=compute_metrics,
        data_collator=DataCollatorWithPadding(tokenizer)
    )

    pre_metrics = trainer.evaluate()
    trainer.train()
    post_metrics = trainer.evaluate()
    learning_curve_data[model_name] = trainer.state.log_history

    # Measure efficiency metrics
    print("   ‚è±Ô∏è Measuring efficiency...")
    inference_time = measure_inference_time(model, tokenizer, sample_texts_for_inference, device)
    model_size = measure_model_size(model)

    # Make predictions and calculate confusion matrix
    preds = trainer.predict(split["test"])
    y_true = preds.label_ids
    y_pred = np.argmax(preds.predictions, axis=1)

    cm = confusion_matrix(y_true, y_pred)
    confusion_matrices[model_name.split("/")[-1]] = cm

    # Calculate per-class metrics
    per_class_f1 = f1_score(y_true, y_pred, average=None, zero_division=0)
    per_class_prec = precision_score(y_true, y_pred, average=None, zero_division=0)
    per_class_rec = recall_score(y_true, y_pred, average=None, zero_division=0)

    short_name = model_name.split("/")[-1]
    for i, score in enumerate(per_class_f1):
        per_class_metrics_list.append({"Model": short_name, "Class": reverse_label_map[i], "F1 Score": score})
    for i, score in enumerate(per_class_prec):
        per_class_precision_list.append({"Model": short_name, "Class": reverse_label_map[i], "Precision": score})
    for i, score in enumerate(per_class_rec):
        per_class_recall_list.append({"Model": short_name, "Class": reverse_label_map[i], "Recall": score})

    # Store summary metrics
    summary_metrics.append({
        "Model": short_name,
        "Pre-Train F1": pre_metrics["eval_f1_weighted"],
        "Post-Train F1": post_metrics["eval_f1_weighted"],
        "Post-F1 (macro)": post_metrics["eval_f1_macro"],
        "Post-Precision (macro)": post_metrics["eval_precision_macro"],
        "Post-Recall (macro)": post_metrics["eval_recall_macro"],
        "Accuracy": post_metrics["eval_accuracy"],
        "Inference Time (ms)": inference_time,
        "Model Size (MB)": model_size
    })

    # Clean up GPU memory
    del model, trainer
    torch.cuda.empty_cache()

# Display summary of all model results
results_df = pd.DataFrame(summary_metrics).sort_values("Post-F1 (macro)", ascending=False)
display(results_df.round(4))

# PLOT 1: Learning Gain - Pre-trained vs Fine-tuned F1 Score
plot_df = results_df.melt(id_vars=["Model"], value_vars=["Pre-Train F1", "Post-Train F1"], var_name="Stage", value_name="F1 Score")
plt.figure(figsize=(12, 6))
sns.barplot(data=plot_df, x="Model", y="F1 Score", hue="Stage")
plt.title("Learning Gain: Pre-trained vs Fine-tuned", fontsize=14, fontweight="bold")
plt.xticks(rotation=15)
plt.ylim(0, 1.05)
plt.show()

# PLOT 2: Efficiency Frontier - Performance vs Efficiency
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=results_df,
    x="Inference Time (ms)",
    y="Post-Train F1",
    size="Model Size (MB)",
    sizes=(200, 1500),
    hue="Model",
    palette="deep",
    alpha=0.8
)
plt.title("Efficiency Frontier: Modern SOTA vs Standard Baselines", fontsize=14, fontweight='bold')
plt.xlabel("Speed: Inference Latency (ms per sentence) ‚Üê Lower is Better")
plt.ylabel("Performance: F1 Score (Accuracy) ‚Üí Higher is Better")
plt.grid(True, linestyle='--', alpha=0.6)

# Add labels to scatter plot bubbles
for i in range(results_df.shape[0]):
    short_name = results_df["Model"].iloc[i].split("/")[-1]
    plt.text(
        results_df["Inference Time (ms)"].iloc[i]+0.2,
        results_df["Post-Train F1"].iloc[i],
        short_name,
        fontsize=10,
        weight='bold'
    )
plt.tight_layout()
plt.show()

# PLOT 3: Macro-F1 Comparison of Models
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df, x="Model", y="Post-F1 (macro)")
plt.title("Model Comparison (Macro-F1)", fontsize=14, fontweight="bold")
plt.xticks(rotation=15)
plt.ylim(0, 1.05)
plt.show()

# PLOT 4: Confusion Matrices for each model
print("\n--- Confusion Matrices ---")
for model_name, cm in confusion_matrices.items():
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", xticklabels=label_list, yticklabels=label_list, cmap="Blues")
    plt.title(f"Confusion Matrix ‚Äì {model_name}")
    plt.ylabel("True Label")
    plt.xlabel("Predicted Label")
    plt.tight_layout()
    plt.show()

# PLOT 5: Per-Class F1 Heatmap
per_class_df = pd.DataFrame(per_class_metrics_list)
if not per_class_df.empty:
    plt.figure(figsize=(12, 5))
    sns.heatmap(per_class_df.pivot(index="Model", columns="Class", values="F1 Score"), annot=True, cmap="Blues", fmt=".2f")
    plt.title("Class-wise F1 Score")
    plt.tight_layout()
    plt.show()

# PLOT 6: Per-Class Precision & Recall Heatmaps
print("\n--- Per-Class Precision & Recall ---")
prec_df = pd.DataFrame(per_class_precision_list)
rec_df = pd.DataFrame(per_class_recall_list)

if not prec_df.empty:
    plt.figure(figsize=(12, 5))
    sns.heatmap(prec_df.pivot(index="Model", columns="Class", values="Precision"), annot=True, cmap="Greens", fmt=".2f")
    plt.title("Class-wise Precision Score")
    plt.tight_layout()
    plt.show()

if not rec_df.empty:
    plt.figure(figsize=(12, 5))
    sns.heatmap(rec_df.pivot(index="Model", columns="Class", values="Recall"), annot=True, cmap="Reds", fmt=".2f")
    plt.title("Class-wise Recall Score")
    plt.tight_layout()
    plt.show()

# PLOT 7: Learning Curves (Loss and F1) for each model
print("\n--- Learning Curves ---")
for model_name, logs in learning_curve_data.items():
    short_name = model_name.split("/")[-1]
    train_losses, eval_f1, epochs = [], [], []
    for log in logs:
        if 'loss' in log:
            train_losses.append(log['loss'])
            epochs.append(log['epoch'])
        elif 'eval_f1_weighted' in log:
            eval_f1.append(log['eval_f1_weighted'])

    plt.figure(figsize=(14, 4))
    plt.subplot(1, 2, 1)
    if train_losses: plt.plot(epochs, train_losses, marker='o'); plt.title(f'{short_name} - Train Loss')
    plt.subplot(1, 2, 2)
    if eval_f1: plt.plot(range(1, len(eval_f1)+1), eval_f1, marker='x', color='orange'); plt.title(f'{short_name} - Eval F1')
    plt.show()

print("\n‚úÖ DONE. All visualizations restored.")
```
